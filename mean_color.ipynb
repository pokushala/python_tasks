```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

img = plt.imread('ogon.jpg')
```

```python
plt.imshow(img)
```

```python
mean_clr = np.mean(img, axis=(0,1))/255
```

```python
plt.imshow([[mean_clr]])
```

```python
mean_img = np.full((img.shape), mean_clr)
```

```python
plt.imshow(mean_img)
```

```python
plt.figure(figsize=(20, 6))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.imshow(mean_img)
```
