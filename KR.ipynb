```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

img = plt.imread('maxresdefault.png')
r_layer = img[:,:,0]
g_layer = img[:,:,1]
b_layer = img[:,:,2]
```

```python
plt.imshow(img)
```

```python
plt.figure(figsize=(20, 3))
plt.subplot(1, 3, 1)
plt.imshow(r_layer, cmap='Reds')
plt.subplot(1, 3, 2)
plt.imshow(g_layer, cmap='Greens')
plt.subplot(1, 3, 3)
plt.imshow(b_layer, cmap='Blues')
plt.show()
```

```python
# отражения
plt.imshow(np.flip(img, axis=0))
plt.show()
plt.imshow(np.flip(img, axis=1))
plt.show()
```

```python
# повороты
plt.imshow(np.rot90(img, 1, axes=(0,1)))
plt.show()
plt.imshow(np.rot90(img, 2, axes=(0,1)))
plt.show()
plt.imshow(np.rot90(img, 3, axes=(0,1)))
plt.show()
```

```python
# вычитание
plt.imshow(r_layer-g_layer)
```

```python
# размеры изображений (высота х ширина х количество каналов)
img.shape
```

```python
h, w, _ = img.shape
print(h,w)
```

```python
# выделение подизображения
plt.imshow(img[0:img.shape[0]//2, 0:img.shape[1]//2])
plt.show()
plt.imshow(img[img.shape[0]//2:, :img.shape[1]//2])
plt.show()
plt.imshow(img[:img.shape[0]//2, img.shape[1]//2:])
plt.show()
plt.imshow(img[img.shape[0]//2:, img.shape[1]//2:])
plt.show()
```

```python
# изменим подизображение
img2 = img.copy()
img2[:h//2,:w//2] = np.rot90(img[:h//2,:w//2],2,(0,1))
plt.imshow(img2)
plt.show()

img2 = img.copy()
img2[:h//2,:w//2] = np.flip(img[:h//2,:w//2], 0)
plt.imshow(img2)
plt.show()

img2 = img.copy()
img2[:,:w//2] = np.flip(img[:,:w//2], 1)
plt.imshow(img2)
plt.show()
```
